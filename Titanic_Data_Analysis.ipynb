```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
dt=pd.read_csv('/content/Titanic-Dataset.csv')
dt
```

```python
dt['Age'].fillna(dt['Age'].median(), inplace=True)
dt['Embarked'].fillna(dt['Embarked'].mode()[0], inplace=True)
dt.drop('Cabin', axis=1, inplace=True)
dt = pd.get_dummies(dt, columns=['Sex', 'Embarked'], drop_first=True)
dt = pd.get_dummies(dt, columns=['Name'], drop_first=True)
dt = pd.get_dummies(dt, columns=['Ticket'], drop_first=True)
```

```python
x = dt[['PassengerId','Pclass','Sex_male','Age','SibSp','Parch','Fare','Embarked_Q','Embarked_S']]
y = dt['Survived']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Columns in the dataset:", dt.columns)
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(y_test, y_pred))
print("Columns in the dataset:", dt.columns)
```
